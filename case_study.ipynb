# importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# read the dataset to the python environment

```python
cars='cars_data.csv'
df=pd.read_csv(cars)
```

# printing dataset

```python
df
```

# check for the null values present in the dataset

```python
df.info()
j=0
for i in df:
    j+=df[i].isnull().sum()
print("\n\nThere are",j,"null values present in the dataset")

    
```

# plot a bar graph of male vs female buyers participated in the sales

```python
plt.figure(figsize=(10,30))
sns.set_theme(style='whitegrid')
sns.countplot(x='Buyer Gender',data=df)
plt.title('Male Vs Female Buyer',size=20)

```

# Bar graph with color of car wise buyers

```python
plt.figure(figsize=(20,40))
sns.set_theme(style='darkgrid')
sns.countplot(x='Buyer Gender',hue='Color',data=df)
plt.title('Male Vs Female Buyer based on color',size=20)
```
